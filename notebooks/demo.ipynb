```python
import matplotlib.pyplot as plt
import numpy as np
from src.core.entropy_gravity import calculate_gradient

# Тестовые узлы судьбы (DN)
nodes = [{'x': 5, 'y': 5, 'm': 100}, {'x': -5, 'y': -5, 'm': 150}, {'x': 0, 'y': 8, 'm': 50}]
subject = {'x': 0, 'y': 0}

# Вычисление вектора
grad = calculate_gradient(nodes, subject)
print(f"Градиент в точке (0,0): {grad}")

# Визуализация
plt.quiver(0, 0, grad['x'], grad['y'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-10, 10); plt.ylim(-10, 10)
plt.title('Entropy Gradient Field (∇Sinf)')
plt.show()
```
